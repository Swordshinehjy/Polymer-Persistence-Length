```python
import polymer_pl as pp
import scipy.constants as sc
import numpy as np

temperature = 293  # K
# in the original paper kTval=2.43652, 293 K
l = [7.01402,1.46690,6.82763, 1.46688]
Angle = np.array([1.73652,-1.77038, 6.74532, -6.76756])
rotation = np.array([0, 1, 0, 1])
labels = {
    1: {'loc': 'PBN-27.txt', 'color': 'b'},
}
```

```python
pBN27 = pp.PolymerPersistence(l, Angle, temperature, rotation, labels, fittting_method='cosine', cosine_deg=14)
pBN27.plot_dihedral_potentials()
```

```python
pBN27.plot_correlation_function(n_repeat_units=20,
                                n_samples=150000,
                                start_idx=1,
                                end_idx=10)  # Monte Carlo method
# original paper: 2.98472*10^6
```
