```python
import polymer_pl as pp
import scipy.constants as sc
import numpy as np

temperature = 293  # K
# in the original paper kTval=2.43652, 293 K
l = [7.04686,1.45969,6.39989, 1.45865]
Angle = np.array([3.69074,-2.89585, -11.34954, -11.70759])
rotation = np.array([0, 1, 0, 1])
labels = {
    1: {'loc': 'P2.txt', 'color': 'b'},
}
```

```python
p2 = pp.PolymerPersistence(l, Angle, temperature, rotation, labels, fittting_method='cosine', cosine_deg=14)
p2.plot_dihedral_potentials()
```
