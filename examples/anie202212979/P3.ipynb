```python
import polymer_pl as pp
import scipy.constants as sc
import numpy as np

temperature = 293  # K
# in the original paper kTval=2.43652, 293 K
l = [7.04214, 1.46050, 2.52510 ,1.44450, 2.52454,1.46034]
Angle = np.array([4.24601,-4.05624,14.57071, 15.1829, -15.32605, -14.62311])
rotation = np.array([0, 2, 0, 1, 0, 2])
labels = {
    1: {'loc': 'T-T.txt', 'color': 'b'},
    2: {'loc': 'BNBP-T.txt', 'color': 'm'},
}
```

```python
p3 = pp.PolymerPersistence(l, Angle, temperature, rotation, labels, fittting_method='cosine', cosine_deg=14)
p3.plot_dihedral_potentials()
```

```python
p3.plot_correlation_function(n_repeat_units=20,
                                n_samples=150000,
                                start_idx=1,
                                end_idx=10)  # Monte Carlo method
# original paper: 7.16071
```
