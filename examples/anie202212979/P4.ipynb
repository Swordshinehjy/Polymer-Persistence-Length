```python
import polymer_pl as pp
import scipy.constants as sc
import numpy as np

temperature = 293  # K
# in the original paper kTval=2.43652, 293 K
l = [7.02063, 1.46206, 2.46891,1.44198, 2.46894, 1.46247]
Angle = np.array([4.83823,-4.70901, 12.20965, 11.98955, -11.93443, -12.18497])
rotation = np.array([0, 2, 0, 1, 0, 2])
labels = {
    1: {'loc': 'Tz-Tz.txt', 'color': 'b'},
    2: {'loc': 'BNBP-Tz.txt', 'color': 'm'},
}
```

```python
p4 = pp.PolymerPersistence(l, Angle, temperature, rotation, labels, fittting_method='cosine', cosine_deg=14)
p4.plot_dihedral_potentials()
```

```python
p4.plot_correlation_function(n_repeat_units=20,
                                n_samples=150000,
                                start_idx=1,
                                end_idx=10)  # Monte Carlo method
# original paper: 9.01015
```
