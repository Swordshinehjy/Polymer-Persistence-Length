```python
import numpy as np
import polymer_pl as pp
import matplotlib.pyplot as plt

temperature = 700  # K
l1 = [9.496, 1.450, 2.497, 1.450, 2.497, 1.450]
Angle1 = np.array([-3.76, 3.76, 17.15, 15.19, -15.19, -17.15])
rotation1 = np.array([0, 1, 0, 2, 0, 1])
l2 = [9.483, 1.450, 2.497, 1.450, 2.497, 1.450]
Angle2 = np.array([6.69, 6.69, 17.15, 15.19, 15.19, 17.15])
rotation2 = np.array([0, 1, 0, 3, 0, 1])
l3 = [8.702, 1.450, 2.497, 1.450, 2.497, 1.450]
Angle3 = np.array([11.12, -11.12, 17.07, 15.19, -15.19, -17.07])
rotation3 = np.array([0, 1, 0, 2, 0, 1])
l4 = [8.462, 1.450, 2.497, 1.450, 2.497, 1.450]
Angle4 = np.array([-17.99, 17.99, 17.08, 15.19, -15.19, -17.08])
rotation4 = np.array([0, 1, 0, 2, 0, 1])
labels1 = {
    1: {'loc': 'NDT1-MeT.txt', 'color': 'b'},
    2: {'loc': 'MeT-MeT-TailTail.txt', 'color': 'm'},
}
labels2 = {
    1: {'loc': 'NDT2-MeT.txt', 'color': 'b'},
    3: {'loc': 'MeT-MeT-TailTail-inv.txt', 'color': 'c'},
}
labels3 = {
    1: {'loc': 'NDT3-MeT.txt', 'color': 'b'},
    2: {'loc': 'MeT-MeT-TailTail.txt', 'color': 'm'},
}
labels4 = {
    1: {'loc': 'NDT4-MeT.txt', 'color': 'b'},
    2: {'loc': 'MeT-MeT-TailTail.txt', 'color': 'm'},
}
```

```python
ndt1 = pp.PolymerPersistence(l1, Angle1, temperature, rotation1, labels1)
ndt2 = pp.PolymerPersistence(l2, Angle2, temperature, rotation2, labels2)
ndt3 = pp.PolymerPersistence(l3, Angle3, temperature, rotation3, labels3)
ndt4 = pp.PolymerPersistence(l4, Angle4, temperature, rotation4, labels4)
ndt1.report()
ndt2.report()
ndt3.report()
ndt4.report()
```

```python
pp.compare_persistence_results([ndt1, ndt2, ndt3, ndt4], ['NDT1', 'NDT2', 'NDT3', 'NDT4'], np.linspace(300, 700, 41))
```
